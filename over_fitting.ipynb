```python
import pandas as pd
```

```python
df = pd.read_csv("./data/sonar3.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
X = df.iloc[:,0:60]
y = df.iloc[:,60]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(24,input_dim=60,activation='relu'),
    Dense(10,activation='relu'),
    Dense(1,activation='sigmoid'),
])
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# history = model.fit(X,y,epochs=200,batch_size=10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,shuffle=True)
```

```python
history = model.fit(X_train,y_train,epochs=200,batch_size=10)
```

```python
score = model.evaluate(X_test,y_test)
print(score[1])
```

```python
model.save("./data/model/Sequential_model.h5")
```

```python
from sklearn.model_selection import KFold
```

```python
k=5
kfold = KFold(n_splits=k,shuffle=True)
acc_score=[]
```

```python
for train_index,test_index in kfold.split(X):
  X_train,X_test = X.iloc[train_index,:],X.iloc[test_index,:]
  y_train,y_test = y.iloc[train_index],y.iloc[test_index]

  model_ff = Sequential([
    Dense(24,input_dim=60,activation='relu'),
    Dense(10,activation='relu'),
    Dense(1,activation='sigmoid'),
  ])
  model_ff.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

  history = model_ff.fit(X_train,y_train,epochs=150,batch_size=10)

  accuracy = model_ff.evaluate(X_test,y_test)
  print(f"loss = {accuracy[0]}")
  print(f"acc = {accuracy[1]}")
```
